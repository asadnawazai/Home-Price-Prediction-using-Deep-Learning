```python
data['ocean_proximity'] = data['ocean_proximity'].map({'<1H OCEAN':0,'INLAND':1,'NEAR OCEAN':2,'NEAR BAY':3,'ISLAND':4})
```

```python
x = data.drop(columns = ['median_house_value'])
y = data['median_house_value']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
min = MinMaxScaler()
x_train_norm = min.fit_transform(x_train)
x_test_norm = min.transform(x_test)
```

```python
x_train_norm
```

```python
x_test_norm
```

```python
x_train_new = pd.DataFrame(x_train_norm, columns=x_train.columns)
x_test_new = pd.DataFrame(x_test_norm,columns=x_test.columns)
```

```python
x_train_new.head()
```

```python
x_test_new.head()
```

# Model Training and structure

```python
model = Sequential([
    Dense(1000, input_shape =[x_train_new.shape[1],], activation='relu'),
    Dense(500, activation='relu'),
    Dense(250,activation='relu'),
    Dense(1,activation='linear')
])
```

```python
model.summary()
```

```python
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
es = EarlyStopping(monitor='val_loss',mode = 'min', patience = 5, restore_best_weights=True)
```

```python
history = model.fit(x_train_new,y_train, validation_data=(x_test_new,y_test), callbacks=[es], epochs=100,batch_size=50, verbose=1)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,mean_squared_log_error
y_pred = model.predict(x_test_new)
```

```python
print("mae :", mean_absolute_error(y_test,y_pred))
print("mse :", mean_squared_error(y_test,y_pred))
print("mae :", mean_squared_log_error(y_test,y_pred))
print("score :", r2_score(y_test,y_pred))
```

```python
score = r2_score(y_pred,y_test)
print(score)
```
